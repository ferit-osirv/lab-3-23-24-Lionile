```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r--', label='sin(x)')
plt.legend()
plt.title('Sinus')
plt.xlabel('x')
plt.ylabel('y')
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
axs[0, 0].plot(x, y)
axs[0, 1].plot(x, y, 'r--', label='sin(x)')
cos = np.cos(x)
axs[1, 0].plot(x, cos, label='cos(x)')
axs[1, 1].plot(x, cos, 'g--', label='cos(x)')
```

```python
x_noisy = np.linspace(0, 2*np.pi, 100)
y_noisy = np.sin(x_noisy) + np.random.normal(0, 0.2, 100)
plt.scatter(x_noisy, y_noisy, label='sin(x) + noise')
```

```python
plt.scatter(x_noisy, y_noisy, label='sin(x) + noise')
plt.plot(x, y, 'r--', label='sin(x)')
plt.legend()
```
